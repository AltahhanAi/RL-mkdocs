As the robot is launched in the gazebo environment, we can see what topics it is publishing and subscribing to. In order to do that, we can use

```python
ros2 topic list   # this tells the list of topics

/clock
/cmd_vel
/imu
/joint_states
/odom
/parameter_events
/performance_metrics
/robot_description
/rosout
/scan
/tf
/tf_static
```

```python
ros2 topic info /odom  # this tell the more information about the topic

Type: nav_msgs/msg/Odometry
Publisher count: 1
Subscription count: 0
```

```python
ros2 interface show nav_msgs/msg/Odometry  # it gives information about different msgs type

# This represents an estimate of a position and velocity in free space.
# The pose in this message should be specified in the coordinate frame given by header.frame_id
# The twist in this message should be specified in the coordinate frame given by the child_frame_id

# Includes the frame id of the pose parent.
std_msgs/Header header

# Frame id the pose points to. The twist is in this coordinate frame.
string child_frame_id

# Estimated pose that is typically relative to a fixed world frame.
geometry_msgs/PoseWithCovariance pose

# Estimated linear and angular velocity relative to child_frame_id.
geometry_msgs/TwistWithCovariance twist
```

Now, lets send some commands to robot using the keyboard press on the geometry_msgs.msg/Twist

Open another terminal, and source it first. Then, select the model of the trutlebot3

```python
export TURTLEBOT3_MODEL=burger
```

Now, lets run the command to bring the teleop keyboard

```python
ros2 run turtlebot3_teleop teleop_keyboard
```

```python
Control Your TurtleBot3!
---------------------------
Moving around:
        w
   a    s    d
        x

w/x : increase/decrease linear velocity (Burger : ~ 0.22, Waffle and Waffle Pi : ~ 0.26)
a/d : increase/decrease angular velocity (Burger : ~ 2.84, Waffle and Waffle Pi : ~ 1.82)

space key, s : force stop

CTRL-C to quit
```

Use the above keys to move the robot around in the environment. One can see the robot moving and turning and also see the values of the laser scan changing. 

Pressing the key strokes will increase the linaer velocity and robot will keep on moving in with the velocity you set by pressing key strokes.
